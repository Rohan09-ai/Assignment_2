**Ques 5. Why might you use a set instead of a list in Python?**

**ans->**A set is an unordered collection of unique elements, while a list is an ordered collection that can contain duplicate elements. You might prefer using a set over a list in the following cases:

**To Ensure Uniqueness:**
Sets automatically eliminate duplicate values. This is useful when you want to store only distinct elements without manually checking for duplicates.

**Faster Membership Testing:**
Sets provide faster performance for checking whether an element exists (using hashing), typically in constant time O(1), whereas lists take linear time O(n).

**Use of Set Operations:**
Sets support mathematical operations like union, intersection, difference, and symmetric difference, which are not directly available with lists.

**Improved Performance in Certain Tasks:**
For large collections of data where order is not important and duplicates are not allowed, sets can be more efficient in terms of both speed and memory.

**Ques 8.What is a hash table, and how does it relate to dictionaries in Python?**

Relation to Dictionaries in Python:
In Python, a dictionary is an implementation of a hash table.

->Keys in a dictionary are passed through a hash function to determine their location in memory.

**Ques 10.Explain why strings are immutable in Python?**

**ans->**Strings are immutable in Python, meaning once a string is created, it cannot be changed. Any operation that seems to modify a string actually creates a new string object.

**Reasons Why Strings Are Immutable:**
->Security and Reliability:

**Ques 14.How does the “in” keyword work differently for lists and dictionaries?**

**Ques 17. Describe the time complexity of accessing elements in a dictionary?**

**ans->** The time complexity of accessing an element in a dictionary is:

->Average Case: O(1) (Constant time)

-> Worst Case: O(n) (Very rare)

Explanation:
Python dictionaries are implemented using a data structure called a hash table. Here's how it works:

->Step-by-step:

1.Key is hashed (converted to a number using hash()).

2.This hash is used to calculate an index.

3.The value is retrieved directly from that index.

Because of this, accessing a value using a key is very fast — it doesn’t need to go through every element like a list.


```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
print(my_dict['b'])  # O(1)
```


**Ques 19.Why are dictionaries considered unordered, and how does that affect data retrieval?**

**ans->** Why were dictionaries considered unordered?


*   Dictionaries use a hash table to store key-value pairs.
*   This structure is optimized for quick access using keys, not for maintaining order.

* The location of each key-value pair in memory depends on the hash value of the key, not on when or where it was added.

**Effect on Data Retrieval:**


*   You could not rely on the order of items.
*  Two dictionaries with the same items might show them in different orders when printed or iterated.
*  we access values using keys, not indexes like lists.











**Ques 20. Explain the difference between a list and a dictionary in terms of data retrieval.**

**ans->The main difference between a list and a dictionary in terms of data retrieval lies in how elements are accessed:**

**-> List: Index-Based Retrieval:**


1.   In a list, elements are accessed by their position (index).
2.  Indexes start from 0.

3.  Retrieval is position-dependent.
4. Time Complexity:
Access by index:O(1) (very fast)

Example:

```python
my_list = ['apple', 'banana', 'mango']
print(my_list[1])   # Output: banana
```

**Dictionary: Key-Based Retrieval:**



1. In a dictionary, elements are
accessed using a Key.  
2.  Keys are unique, and each key maps to a specific value.

3. Retrieval is label-based, not position-based.

4.Time Complexity:
Access by key: O(1) (average case, very fast)



Example:

```python
my_dict = {'fruit1': 'apple', 'fruit2': 'banana'}
print(my_dict['fruit2'])   # Output: banana
```
